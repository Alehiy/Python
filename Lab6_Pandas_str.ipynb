```python
import pandas as pd
```

```python
df = pd.read_csv (r'C:\Users\alehi\Downloads\books.csv')
df
```

```python
#Створіть стовпці з кількістю голосних в імені автора.
df2=df
df2[['last_name','first_name']] = df['Author'].loc[df['Author'].str.split().str.len() == 2].str.split(', ',expand=True)
df2[['vowel_count']] = (df.first_name.str.extractall('(?i)(?P<vowels>[aeiou])').groupby(level=0).count())
df2
```

```python
#Знайдіть авторів із прізвищами з 3 голосними поспіль.
##Таких нет, поэтому нашел с 2 гласными подряд
df2['last_name'].str.lower()
#df3=df2['last_name'].dropna().str.findall(r'^.*[aeiou][aeiou][aeiou].*$')
df3=df2['last_name'].dropna().str.findall(r'^.*[aeiou][aeiou].*$')
df3.loc[(df3.str.len()>0)]
```

```python
#Створіть стовпець і внесіть туди 1, якщо в назві є цифри, і 0 у протилежному випадку
df2.loc[df2['Title'].str.contains('\\d', regex=True), 'cont_num'] = '1'
df2.loc[~df2['Title'].str.contains('\\d', regex=True), 'cont_num'] = '0'
pd.set_option('display.max_rows', df2.shape[0]+1)
df2
```

```python
# Знайдіть найдовший заголовок у наборі даних.
df2.loc[df2['Title'].str.len()==df2['Title'].str.len().max()]
```

```python
#Знайдіть книги, написані автором, ім’я якого починається на „A"
df2.loc[(df2['first_name'].str.startswith("A")==True)]
```

```python
#Знайдіть книги, видані видавництвами, у назві яких 2 слова
df2.loc[df2['Publisher'].str.contains(' ',na=False, regex=True)]
```

```python
# Знайдіть авторів із 3-м «у» у прізвищі
df3 = df2['last_name'].dropna().str.findall(r'^..y.*$')
df3.loc[(df3.str.len()>0)]
```
